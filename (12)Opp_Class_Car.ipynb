```python
class Car:
    
    def __init__(a,seat,bettery,engine,deshborad):
        a.seat = seat
        a.bettery = bettery
        a.Engine = engine
        a.deshborad = deshborad
    def __str__(a):
        return f"Bettery is mode up of {a.bettery} .Seat is manfuction in soft leather {a.seat} Dashboard is made up of iron steel metal{a.deshborad} Engine is made of {a.Engine}"

class Engine:
    def __init__(a,company,horse_power,fuel,bettery):
        a.company = company
        a.horse_power = horse_power
        a.fuel = fuel
        a.bettery = bettery
    def  __str__(a):
        return f"Engine is chekin by {a.company} .Engine in this engine almost {a.horse_power}.The fuel filling capacity level is upto {a.fuel}Bettery is made upto {a.bettery}"



class Base_car:
    def __init__(a,seat,bettery,engine,deshborad):
        super().__init__(seat,bettery,engine,deshborad)

    def __str__(a):
        return f"Bass car:{super().__str__()}"

class LuxuryCar:
    def __init__(a,seat,bettery,engine,deshborad):
        super().__init__(seat,bettery,engine,deshborad)

    def __str__(a):
        return f"Luxury Car:{super().__str__()}"

class ElectronicEngine(Engine):
    def __init__(a,company,horse_power,bettery,fuel):
        super().__init__(company,horse_power,fuel,bettery)
        

    def __str__(a):
        return f"Electronicngine:{super().__str__()}"

class petrolEngine(Engine):
    def __init__(a, company, horse_power, fuel,buttery,):
        super().__init__(company, horse_power, fuel,buttery)

    def __str__(a):
        return f"PetrolEngine:{super().__str__()}"
    
class Hybirdine(Engine):
    def __init__(a, company, horse_power, fuel,bettery):
        super().__init__(company, horse_power, fuel,bettery)
        

    def __str__(a):
        return f"Hybirdine:Bettery:{a.bettery} {super().__str__()}"
        
Engine1 = Engine("k G F",1000,"YES","oska")
print(Engine1)
CAR1 = Car("leather","oska",Engine1,"sample")
print(CAR1)

E = ElectronicEngine("KAG",2000,"yes","fuel")
print(E)

            #Inharetance

class Employeed:
    def __init__(a,name,last_name,ssn):
        a.name = name
        a.last_name =last_name
        a.ssn = ssn
    

    def getname(a):
        return a.name

    def getlast_name(a):
        return a.last_name

    def getssn(a):
        return a.ssn


    def __str__(a):
        return f"Employ name is {a.name} last name is {a.last_name} Social Sekarotey number {a.ssn}"


class CommissionEmployee(Employeed):
    def __init__(a, name, last_name, ssn,Commission_rate,gross_rate):
        super().__init__(name, last_name, ssn)
        a.commission_rate =Commission_rate
        a.gross_rate = gross_rate

    def maulteply (a):
        return a.commission_rate*a.gross_rate
    def __str__(a):
        return super().__str__() + f'commission rate: {a.commission_rate}\n Gross rate: {a.gross_rate}'


class HourlyEmployee(Employeed):
    def __init__(a, name, last_name, ssn,hourly_worked,hours_wage):
        super().__init__(name, last_name, ssn)
        a.set_hours(hourly_worked)
        a.set_wage(hours_wage)
    
    def set_wage(a,hourly_wage):

k=CommissionEmployee("ABU","BAKAR","03022",0.56,20000)

print(k.getname())
print(k.getlast_name())
print(k.getssn())
print(k.maulteply())
print(k)





class Employee:
    def __init__(self, first_name, last_name, ssn):
        self.first_name = first_name
        self.last_name = last_name
        self.ssn = ssn
    
    def get_first_name(self):
        return self.first_name
    
    def get_last_name(self):
        return self.last_name
    
    def get_social_security_number(self): 
        return self.ssn
    
    def __str__(self):
        return f'Name: {self.first_name} {self.last_name}\nSSN: {self.ssn}'

class CommissionEmployee(Employee):
    def __init__(self, first_name, last_name, ssn, commission_rate, gross_sales):
        super().__init__(first_name, last_name, ssn)
        self.commission_rate = commission_rate
        self.gross_sales = gross_sales
    
    def earnings(self):
        return self.commission_rate * self.gross_sales
    
    def __str__(self):
        return super().__str__() + f'\nCommission rate: {self.commission_rate}\nGross sales: {self.gross_sales}'



class HourlyEmployee(Employee):
    def __init__(self, first_name, last_name, social_security_number, hourly_wage, hours_worked):
        super().__init__(first_name, last_name, social_security_number)
        self.set_wage(hourly_wage)
        self.set_hours(hours_worked)

    def set_wage(self, hourly_wage):
        if hourly_wage < 0:
            raise ValueError("Wage cannot be negative.")
        self._wage = hourly_wage

    def get_wage(self):
        return self._wage

    def set_hours(self, hours_worked):
        if hours_worked < 0 or hours_worked > 168:
            raise ValueError("Hours worked must be between 0 and 168.")
        self._hours = hours_worked

    def get_hours(self):
        return self._hours

    def earnings(self):
        if self._hours <= 40:
            return self._wage * self._hours
        else:
            return 40 * self._wage + (self._hours - 40) * self._wage * 1.5

    def __str__(self):
        return super().__str__() + f"\nWage: {self._wage}\nHours worked: {self._hours}\nEarnings: {self.earnings()}"



            # Create a HourlyEmployee object
he = HourlyEmployee('John', 'Doe', '123-45-6789', 10, 50)

            # Test the methods
print(he.get_first_name()) # Output: John
print(he.get_last_name()) # Output: Doe
print(he.get_social_security_number()) # Output: 123-45-6789
print(he.get_wage()) # Output: 10
print(he.get_hours()) # Output: 5
print(he.earnings()) # Output: 550.0
print(he) # Output: Name: John Doe
          # SSN: 123-45-6789
          # Wage: 10
          # Hours worked: 50
          # Earnings: 550.0

        # Create a CommissionEmployee object
ce = CommissionEmployee('John', 'Doe', '123-45-6789', 0.10, 10000)

                # Test the methods
print(ce.get_first_name()) # Output: John
print(ce.get_last_name()) # Output: Doe
print(ce.get_social_security_number()) # Output: 123-45-6789
print(ce.earnings()) # Output: 1000.0
print(ce) # Output: Name: John Doe
          # SSN: 123-45-6789
          # Commission rate: 0.1
          # Gross sales: 10000.0


```
